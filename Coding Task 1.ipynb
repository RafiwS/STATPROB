```python
from collections import Counter
from scipy.stats import skew, kurtosis

curah_hujan = [ 1.4, 1.2, 15.2, 13.4, 7.7, 0.6, 1, 0, 0, 3.2, 0, 6.2, 0, 37.2, 1.6, 16.1, 25.4, 19.8, 0, 11.4, 4.2, 20.5, 54.5, 1, 6.4, 17.8, 0, 5.2, 
               0, 6.4, 55.9, 4, 5.6, 0.3, 0, 2.1, 1.5, 6.4, 64, 0.5, 0, 2.4, 1, 0, 0, 0, 0, 58, 0, 1, 1, 38.6, 21, 0, 0, 12, 13.2, 0, 9, 4, 100.6, 14.2, 
               6.6, 1, 15.1, 34.9, 1, 1, 0, 0.5, 1, 3.3, 1, 9.4, 40.1, 17.6, 170.6, 0, 16.4, 0, 6.8, 45.5, 68, 18.4, 1, 45.4, 1, 0, 0, 0, 0, 31.6, 0, 2.2, 
               0.5, 33.4, 0, 0, 17.6, 170.6, 0, 16.4, 0, 6.8, 45.5, 68, 18.4, 1, 45.4, 1, 0, 0, 0, 0, 31.6, 0, 2.2, 0.5, 33.4, 0, 0 ]
penjualan_a = [ 0, 27, 458, 20, 7, 0, 4, 350, 1, 240, 13, 169, 0, 18, 1, 206, 10, 11, 200, 0, 9, 305, 360, 373, 34, 0, 0, 220, 22, 9, 208, 17, 
               213, 0, 24, 0, 210, 163, 250, 28, 0, 130, 269, 252, 0, 15, 176, 0, 416, 301, 239, 204, 2, 11, 2, 266, 311, 123, 419, 205, 181, 0, 303, 
               80, 138, 0, 148, 389, 0, 45, 11, 300, 5, 238, 3, 0, 301, 52, 348, 64, 125, 197, 300, 0, 410, 391, 32, 17, 17, 0, 23, 37, 0, 430, 208, 200, 
               0, 143, 0, 301, 52, 348, 64, 125, 197, 300, 0, 410, 391, 32, 17, 17, 0, 23, 37, 0, 430, 208, 200, 0, 143 ]
penjualan_b = [ 519, 310, 139, 76, 0, 423, 244, 325, 340, 39, 524, 333, 341, 0, 517, 37, 144, 60, 221, 138, 412, 163, 170, 0, 370, 105, 428,
               11, 430, 229, 26, 218, 72, 6, 440, 158, 313, 296, 236, 0, 265, 420, 208, 4, 416, 487, 339, 242, 263, 0, 236, 157, 55, 47, 2, 229, 
               67, 7, 71, 28, 166, 57, 255, 24, 302, 38, 177, 341, 530, 103, 512, 155, 0, 141, 83, 118, 184, 367, 332, 59, 271, 209, 136, 519, 66, 
               164, 13, 220, 359, 227, 388, 231, 0, 187, 34, 163, 298, 93, 456, 57, 415, 51, 134, 328, 53, 34, 344, 65, 81, 106, 192, 513, 80, 143, 208, 
               109, 144, 175, 96, 360, 7 ]

def calculate_statistics(data):
    n = len(data)
    data_sorted = sorted(data)

    # mean
    mean = sum(data) / n

    # median
    if n % 2 == 0:
        median = (data_sorted[n // 2 - 1] + data_sorted[n // 2]) / 2
    else:
        median = data_sorted[n // 2]

    # modus
    data_counter = Counter(data)
    mode = data_counter.most_common(1)[0][0]

    # jangkauan
    range_value = max(data) - min(data)

    # jangkauan interquartil
    q1 = data_sorted[int(n * 0.25)]
    q3 = data_sorted[int(n * 0.75)]
    iqr = q3 - q1

    # variasi
    variance = sum((x - mean) ** 2 for x in data) / (n - 1)

    # standar deviasi
    std_dev = variance ** 0.5

    # Skewness
    skewness = skew(data)

    # Kurtosis
    kurt = kurtosis(data)
    
    return {
        'Mean': mean,
        'Median': median,
        'Mode': mode,
        'Range': range_value,
        'IQR': iqr,
        'Variance': variance,
        'Std Dev': std_dev,
        'Skewness': skewness,
        'Kurtosis': kurt
    }

datasets = {
    'Curah Hujan (mm)': curah_hujan,
    'Penjualan A (pcs)': penjualan_a,
    'Penjualan B (pcs)': penjualan_b
}

for name, data in datasets.items():
    stats = calculate_statistics(data)
    print(f"\nStatistik untuk {name}:\n" + "-" * 30)
    for stat_name, value in stats.items():
        print(f"{stat_name:<15}: {value:.2f}") 
```

```python
import pandas as pd
import math

data = [ 1.4, 1.2, 15.2, 13.4, 7.7, 0.6, 1, 0, 0, 3.2, 0, 6.2, 0, 37.2, 1.6, 16.1, 25.4, 19.8, 0, 11.4, 4.2, 20.5, 54.5, 1, 6.4, 17.8, 0, 5.2, 
        0, 6.4, 55.9, 4, 5.6, 0.3, 0, 2.1, 1.5, 6.4, 64, 0.5, 0, 2.4, 1, 0, 0, 0, 0, 58, 0, 1, 1, 38.6, 21, 0, 0, 12, 13.2, 0, 9, 4, 100.6, 14.2, 
        6.6, 1, 15.1, 34.9, 1, 1, 0, 0.5, 1, 3.3, 1, 9.4, 40.1, 17.6, 170.6, 0, 16.4, 0, 6.8, 45.5, 68, 18.4, 1, 45.4, 1, 0, 0, 0, 0, 31.6, 0, 2.2, 
        0.5, 33.4, 0, 0, 17.6, 170.6, 0, 16.4, 0, 6.8, 45.5, 68, 18.4, 1, 45.4, 1, 0, 0, 0, 0, 31.6, 0, 2.2, 0.5, 33.4, 0, 0 ]

n = len(data)
print("1. n =", n)

d_min = min(data)
d_max = max(data)
print("2. Dmin =", d_min, ", Dmax =", d_max)

R = d_max - d_min
print("3. R =", R)

k = math.ceil(1 + 3.3 * math.log(n, 10))
print("4. Banyak kelas =", k)

I = math.ceil(R / k)
print("5. Panjang interval kelas =", I)

kelas = []
interval_kelas = []
tepi_kelas = [] 
for i in range(k):
    BAK = d_min + i * I
    BBK = BAK + I
    tepi = BAK - 0.5
    kelas.append(i + 1)
    interval_kelas.append(str(BAK) + "-" + str(BBK))
    tepi_kelas.append(tepi)
tepi_kelas.append(BBK + 0.5)
print("6. Kelas =", kelas)
print("7. Interval kelas =", interval_kelas)
print("8. Tepi kelas =", tepi_kelas)


df = pd.DataFrame(data, columns=["Value"])

df["Range"] = pd.cut(df["Value"], bins=tepi_kelas, labels=interval_kelas, include_lowest=True)
frequency_table = df["Range"].value_counts().sort_index()
relative_frequency = frequency_table / frequency_table.sum() * 100

frequency_distribution = pd.DataFrame(
    {
        "Kelas": kelas,
        "Range": frequency_table.index,
        "Frequency": frequency_table.values,
        "Relative Frequency (%)": relative_frequency,
    }
)
frequency_distribution.reset_index(drop=True, inplace=True)


print("\nTabel Distribusi Frekuensi Curah Hujan")
print(frequency_distribution)
```
